# Axes, Label, legend and grid

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days= [1,2,3,4,5,6,7]
max_t= [45,56,47,51,52,43,54]
min_t =[23,30,32,36,28,27,29]
avg_t =[32, 38, 39.2, 43.5,40.5,40,41.5]
```

```python
plt.xlabel("days")
plt.ylabel("temp")
plt.title("weather report")
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="avg")

plt.legend(loc="best", shadow=True, fontsize ="small")
plt.grid()
```

# Bar chat

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['google', 'amazon', 'msft', 'facebook']
revenue=[90,136,89,27]
profit =[40,2,34,12]
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)
plt.xlabel("company")
plt.ylabel("revenue")
plt.title("revenue chart")
plt.bar(xpos-0.2,revenue, width=0.4, label="revenue")
plt.bar(xpos+0.2, profit, width=0.4, label="profit")
plt.legend()
plt.grid()
```

```python
plt.yticks(xpos,company)
plt.ylabel("company")
plt.xlabel("revenue")
plt.title("revenue chart")
plt.barh(xpos-0.2,revenue, height=0.4 ,label="revenue")
plt.barh(xpos+0.2, profit,height=0.4 , label="profit")
plt.legend()
```

# Histogram

```python
blood_sugar_men = [113,85,90,150,149,88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.xlabel("sugar range")
plt.ylabel("nos of patient")
plt.title("Blood Sugar Analysis")
plt.hist([blood_sugar_men, blood_sugar_women], bins= [80,100,125,150], rwidth=0.95, color=["green", "blue"], 
         label=["men", "women"])
plt.legend()
```

# Pie chart

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["home rent", "food", "phone/internet bill","car", "others" ]
```

```python
plt.pie(exp_vals,labels=exp_labels, radius=1.5,autopct="%.2f%%", shadow=True)
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, radius=1.5,autopct="%.2f%%", shadow=True,explode=[0,0.1,0,0,0])
plt.show()

plt.savefig("piechart.pdf", bbox_inches="tight",pad_inches=2)
```
